```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

First, we will load the data set. These data are the hourly accumulations of rainfall from the beginning of 2017 until the end of 2022 in Robstown, Texas, recorded in millimeters. The head and tail of the data set are shown below.

```python
dat = pd.read_csv('C:\\Downloads\\RBO_hourly.csv')
print(dat.head())
print(dat.tail())
```

Next, we will extract the rainfall column.

```python
rainfall = dat['p01m']
```

Now, we will calculate some statistics of the rainfall data.

```python
print('mean: \t\t', rainfall.mean())
print('median: \t', rainfall.median())
print('max: \t\t', rainfall.max())
print('stdev: \t\t', rainfall.std())
print('kurtosis: \t', rainfall.kurtosis())
print('skewness: \t', rainfall.skew())
```

One interesting statistics of these data is the kurtosis of the distribution of rainfall values. Kurtosis refers to the "tailedness" of a distribution - in other words, how "thick" the tails are of a distribution relative to a normal distribution. A Gaussian distribution has a kurtosis of 3, and a distribution with "thicker" tails will have a kurtosis greater than 3. The kurtosis of the rainfall distribution is ~960, which is indiciative of heavy-tailed data - this is associated with a greater nmber of outliers. Given that most hourly accumulations are 0mm, any noticeable amount of rainfall would be seen as an outlier. 
Additionally, the skewness of the data refers to how symmetric a distribution is. A Gaussian distribution has a skewness of 0, indicating that the distribution is symmetric. The skewness of the rainfall distribution is ~27, which is reflective of a strong right skew.

For reference and context for how skewed the data truly are, two histograms are provided below. The first histogram uses a logarithmic y-axis scale, and the second is a linear scale.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.hist(rainfall, bins = 20)
ax1.set_yscale('log')
ax1.set_xlabel('Hourly Rainfall (mm)')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of Hourly Rainfall, Log Y Scale')

ax2.hist(rainfall, bins = 20)
ax2.set_yscale('linear')
ax2.set_xlabel('Hourly Rainfall (mm)')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of Hourly Rainfall, Linear Y Scale')

plt.show()
```
