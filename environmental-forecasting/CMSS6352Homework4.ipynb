```python
# libraries needed
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

### Matrices (Chapter 2, pages 15-18)

```python
# create matrix A

A = np.matrix([[2, 4, 3, 7], [9, 3, -1, 2], [1, 9, 3, 7], [6, 6, 3, -2]]) # creates matrix A (each vector is a row)
A
```

```python
# calculate column sums

ans = np.sum(A, axis = 0) # sums rows (axis = 0 is column sum)
ans
```

```python
# calculate sum of all elements in A
b = np.sum(sum(A), axis = 1) # sums row of the column sum matrix (axis = 1 is row sum, only one row in matrix so outputs one-element matrix)
b
```

```python
# delete ans from workspace

varbs = dir() # creates list of all variables in workspace
user_varbs = [var for var in varbs if not var.startswith('_')] # creates list of user variables
print(user_varbs) # prints user variables

del ans # deletes ans from workspace

varbs = dir()
user_varbs = [var for var in varbs if not var.startswith('_')] 
print(user_varbs)
```

```python
# replace matrix values

print(A)
A[2, 1] = 30 # replaces 3rd row, 2nd column with 30 (python indexes start at 0)
print(A)
```

```python
# replace multiple elements at once

print(A)
A[2, 0:4] = [1, 3, 3, 5] # replaces 3rd row, 1st - 4th column with [1, 3, 3, 5] (python colon excludes end point)
print(A)
```

```python
# create integer vector

c = list(range(0, 11)) # creates vector from 0 - 10 (integers only)
c
```

```python
# create float vector

c = np.arange(0, 10.5, 0.5) # creates vector from 1 - 10 with step size of 0.5 (linspace specifies number of elements, not step size)
c
```

```python
# vector multiplication

B = np.matrix([[4, 2, 6, 5], [7, 8, 5, 6], [2, 1, -8, -9], [3, 1, 2, 3]]) # creates matrix B
print('B: \n', B)

C = A * B.T # multiplies A by transpose matrix of B
print('C: \n', C)

C = np.multiply(A, B) # multiplies A by B element-wise
print('C: \n', C)
```

### Visualizations (Chapter 2, pages 25-27)

```python
# specify functions

x = np.arange(0, 2*np.pi, 0.1*np.pi) # creates vector from 0 - 2*pi with step size of pi/10
y = np.sin(x) # creates sin(x) vector
```

```python
# basic plot

warnings.filterwarnings("ignore") # ignores warning messages
sns.scatterplot(x = x, y = y) # plots x and y
plt.show() # shows plot
```

```python
# two line plots

y2 = np.cos(x) # creates cos(x) vector

sns.lineplot(x = x, y = y, color = "red", linestyle = "dashed") # plots sin(x) as dashed red line
sns.lineplot(x = x, y = y2, color = "blue") # plots cos(x) as solid blue line
plt.show() # shows plot
```

```python
# create side-by-side plots

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5)) # initializes two figures in one row
sns.lineplot(x = x, y = y, color = "red", linestyle = "dashed", ax = ax1) # plots sin(x) as dashed red line in first figure
sns.lineplot(x = x, y = y2, color = "blue", ax = ax2) # plots cos(x) as solid blue line in second figure
plt.show() # shows plot
```

```python

sns.lineplot(x = x*np.pi, y = y, color = "red", linestyle = "dashed") # multiplication needed to keep axes the same
sns.barplot(x = x, y = y2, color = "blue") # plots cos(x) as blue barplot
plt.show() # shows plot
```

```python
# my first plot

sns.lineplot(x = x, y = y, color = "red", linestyle = "dashed") # plots sin(x) as dashed red line
plt.title("My first plot") # specifies title
plt.xlabel("x-axis") # labels x-axis
plt.ylabel("y-axis") # labels y-axis
plt.show() # shows plot
```
