```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# Creating a Boxplot of NDFD Data 

First, we will load the data into a pandas dataframe.

```python
dat = pd.read_csv('c:\\Users\\radma\\Downloads\\hpccodes\\output2022.csv')
print(dat.head())
print(dat.tail())
```

Next, we will extract the rainfall column.

```python
rainfall = dat['precip']
```

Now, we will produce the boxplot.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.boxplot(rainfall)
ax1.set_xlabel('Hourly Rainfall (mm)')
ax1.set_ylabel('Frequency')
ax1.set_title('Boxplot of Hourly Rainfall')

ax2.violinplot(rainfall)
ax2.set_xlabel('Hourly Rainfall (mm)')
ax2.set_ylabel('Frequency')
ax2.set_title('Violin Plot of Hourly Rainfall, Linear Y Scale')

plt.show()
```

As seen in the box plot above, rainfall is mostly absent. Because of this, most precipitation events are seen as outliers due to how infrequent they are. To get a better understanding of the distribtion, an accompanying violin plot is included as well.
